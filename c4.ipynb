```python
import pandas as pd
import numpy as np
```

```python
data_by_country = pd.read_csv(filepath_or_buffer="data/dataset - data by country.csv")
data_by_country['Users per month'] = data_by_country['Users per month'].astype(str).str.replace(',', '').astype(int)
data_by_country
```

```python
data_by_country['Transactions per User'] = data_by_country['In app Transactions'] / data_by_country['Users per month']
data_by_country.sort_values(by = ['Users per month'], ascending = False).reset_index()
```

```python
data_by_country.describe()
```

```python
# Define conditions and corresponding values
conditions = [
    data_by_country['Users per month'] > 1.592562e+05,
    data_by_country['Users per month'] > 6.981900e+04,
    data_by_country['Users per month'] > 2.815200e+04
]
values = [1, 2, 3]

# Apply conditions using np.select()
data_by_country['Users Tier'] = np.select(conditions, values, default=4)

# Define conditions and corresponding values
conditions = [
    data_by_country['Transactions per User'] > 0.044129,
    data_by_country['Transactions per User'] > 0.036953,
    data_by_country['Transactions per User'] > 0.026723
]
values = [1, 2, 3]

# Apply conditions using np.select()
data_by_country['Transactions Tier'] = np.select(conditions, values, default=4)
```

```python
data_by_country.loc[
    (data_by_country['Transactions Tier'].isin([2, 3])) & (data_by_country['Users Tier'] == 2)
    ].sort_values(by = ['Users per month'], ascending = False)
```
